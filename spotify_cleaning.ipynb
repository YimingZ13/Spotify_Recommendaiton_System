# Correct Formatting

There are unwanted characters like square brackets and quotation marks from artists and genres column in the data frames. Let's remove them.

```python
df['artists'] = df['artists'].str.replace(r"\[|\]|'",'',regex=True)
genre_df['genres'] = genre_df['genres'].str.replace(r"\[|\]|'",'',regex=True)
df_w_genre['genres'] = df_w_genre['genres'].str.replace(r"\[|\]|'",'',regex=True)
```

Now, we have genres with no values.

```python
genre_df[genre_df['genres']=='']
```

Just one data missing genre, I will just fill it in with 'other'.

```python
genre_df[genre_df['genres']==''] = genre_df[genre_df['genres']==''].replace('','other')
```

```python
df_w_genre[df_w_genre['genres']==''].head()
```

```python
len(df_w_genre[df_w_genre['genres']==''])
```

There are quite a lot data missing genres in df_w_genre, for now, let's just fill them with 'other'. Hopefully, it is not affecting the following processes.

```python
df_w_genre[df_w_genre['genres']==''] = df_w_genre[df_w_genre['genres']==''].replace('','other')
```

```python
len(df_w_genre[df_w_genre['genres']==''])
```

Looks like the year and release date column have the same information, let's check.

```python
print(df['year'].unique())
print(f"There are {df['year'].nunique()} unique years.")
```

```python
print(df['release_date'].unique())
print(f"There are {df['release_date'].nunique()} unique release dates.")
```

There are way more release dates than years, obviously we are going to analyze the musics based on monthly or daily information. I will drop the release date column for it is redundant. But first, let's make sure all the values in the year column are indeed equal to years in release date.

```python
# parse out the year from release date, and convert datatype
df['release_date'] = df['release_date'].apply(lambda x: x.split('-')[0]).astype(int)
```

```python
# check if two columns are equal
print(f"There are {(len(df))-((df['year']==df['release_date']).sum())} data do not match.")
```

Great! It is now safe to drop one of these columns.

```python
df.drop(columns='release_date',inplace=True)
```

# Handle Defective Data

```python
# describe the numerical features
df.describe().T
```

One thing to notice, most of these attributes are of value in a relative scale, but tempo or beats per minute is actual numerical representation of a song. It is impossible for a song to have 0 tempo, let's impute these values to the average.

```python
# how many songs with 0 tempo
print(f"There are {len(df[df['tempo']==0])} songs with 0 tempo.")
```

Let's impute these values to the mean instead.

```python
df.loc[df['tempo']==0,'tempo'] = df.loc[df['tempo']>0,'tempo'].mean()
df_w_genre.loc[df_w_genre['tempo']==0,'tempo'] = df_w_genre.loc[df_w_genre['tempo']>0,'tempo'].mean()
```

```python
# sanity check
df['tempo'].describe()
```

# Drop Duplicates

It looks like there are no duplicates in the data frame. But notice the id column in df contains unique information, let's remove this column and check duplicates again.

```python
# check duplicates without id
df = df.drop(columns='id')
```

```python
# create dictionary for the dataframes
df_list = {'track dataframe':df,
           'year dataframe':year_df,
           'genre dataframe':genre_df,
           'track with genre dataframe':df_w_genre}

# print the number of duplicates
for name, i in df_list.items():
    print(f"There are {i.duplicated().sum()}, {round(i.duplicated().sum()/len(i)*100,4)}% duplicated data in {name}.")
```

We have 565 duplicated data in the track dataframe, after excluding the ids, let's look at a duplicated track again to make sure.

```python
df[df.duplicated()].head()
```

```python
# look at duplicates of 'Bitch, Don’t Kill My Vibe'
df[(df['artists']=="Kendrick Lamar") & (df['name']=='Bitch, Don’t Kill My Vibe')]
```

It is safe to just drop them.

```python
# drop duplicates
df = df[~df.duplicated()==True]
```
